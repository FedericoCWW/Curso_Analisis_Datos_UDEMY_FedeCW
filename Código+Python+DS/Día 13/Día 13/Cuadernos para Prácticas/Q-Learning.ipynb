```python
import numpy as np
import random as rand
```

```python
# Definir entorno
dim = (5,5)
init_state = (0,0)
final_state = (4,4)
obstacles = [(1,1), (1,3), (2,3), (3,0)]
actions = [(-1,0), (1,0), (0,-1), (0,1)]
```

```python
states_num = dim[0] * dim[1]
```

```python
actions_num = len(actions)
```

```python
Q = np.zeros((states_num, actions_num))
Q
```

```python
def estado_a_indice(estado):
    return estado[0] * dim[1] + estado[1]
```

```python
print(estado_a_indice((1,2)))
```

```python
alpha = 0.1
gamma = 0.99
epsilon = 0.2
epochs = 10
```

```python
def get_next_state(state, action):
    row, col = divmod(state, 4)  

    if action == 0 and col > 0:        
        col -= 1
    elif action == 1 and col < 3:      
        col += 1
    elif action == 2 and row > 0:     
        row -= 1
    elif action == 3 and row < 3:     
        row += 1

    return row * 4 + col

def elegir_accion(estado):
    if rand.uniform(0,1) < epsilon:
        return rand.choice(range(actions_num))
    else:
        return np.argmax(Q[estado_a_indice(estado)])
    
def aplicar_accion(estado, accion_idx):
    accion = actions[accion_idx]
    nuevo_estado = tuple(np.add(estado,accion) % dim)
    
    if nuevo_estado in obstacles or nuevo_estado == estado:
        return estado, -100, False
    if nuevo_estado == final_state:
        return nuevo_estado, 100, True
    return estado, -1, False
```

```python
for e in range(epochs):
    estado = init_state
    terminado = False
    
    while not terminado:
        idx_state = estado_a_indice(estado)
        idx_accion = elegir_accion(estado)
        nuevo_estado, reward, terminado = aplicar_accion(estado, idx_accion)
        
        idx_nuevo_estado = estado_a_indice(nuevo_estado) if not terminado else None
        
        if terminado:
            target = reward
        else:
            target = reward + gamma * np.max(Q[idx_nuevo_estado])
        
        Q[idx_state, idx_accion] = Q[idx_state, idx_accion] + alpha * (target - Q[idx_state, idx_accion])
        estado = nuevo_estado
```
