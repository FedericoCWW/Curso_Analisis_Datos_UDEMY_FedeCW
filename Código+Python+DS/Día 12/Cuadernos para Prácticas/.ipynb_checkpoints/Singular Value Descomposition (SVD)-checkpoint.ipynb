```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris
```

```python
X = iris.data
```

```python
X_centrado = X - np.mean(X, axis=0)
```

```python
X_centrado
```

```python
U, S, Vt = np.linalg.svd(X_centrado) 
```

```python
k = 2
X_transformado = U[:,:k] * S[:k]
```

```python
especies = iris.target_names
```

```python
especies
```

```python
plt.figure(figsize=(12,6))
for e in range(len(especies)):
    plt.scatter(X_transformado[iris.target == e, 0], X_transformado[iris.target == e, 1], label=especies[e])
    
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.legend()
plt.title("Dataset Iris transformado por SVD");
```
