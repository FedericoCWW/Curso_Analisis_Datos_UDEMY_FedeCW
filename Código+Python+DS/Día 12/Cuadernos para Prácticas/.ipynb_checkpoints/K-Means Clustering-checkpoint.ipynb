```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
data = np.random.rand(100,2)
data
```

```python
modelo = KMeans(n_clusters=5, random_state=42)
modelo.fit(data)
```

```python
centroides = modelo.cluster_centers_
centroides
```

```python
etiquetas = modelo.labels_
etiquetas
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data[:, 0],data[:, 1], c=etiquetas, cmap="viridis", marker='x')
plt.scatter(centroides[:,0],centroides[:,1], c="red", marker="^", s=200, label="Centroides")
plt.title("Visualizacion de K mean Clustering")
plt.grid()
plt.legend()
plt.show();
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.dropna(inplace=True)
```

```python
penguins
```

```python
data2 = penguins[["bill_length_mm", "bill_depth_mm"]]
```

```python
data2
```

```python
modelo2 = KMeans(n_clusters=3, random_state=42)
```

```python
modelo2
```

```python
modelo2.fit(data2)
```

```python
centroides2 = modelo2.cluster_centers_
etiquetas2 = modelo2.labels_
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue=etiquetas2)
plt.scatter(centroides2[:,0],centroides2[:,1], c="red", marker="X", s=200, label="Centroides")
plt.title("Visualizacion de K mean Clustering con pinguinos")
plt.grid()
plt.legend()
plt.show();
```
