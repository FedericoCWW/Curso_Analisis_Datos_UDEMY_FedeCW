```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
import numpy as np
from sklearn.datasets import load_iris
```

```python
data = np.array([[1,-1,4], [7,9,0], [0,0,-3]])
```

```python
# pasamos por Minmaxscaler para pasar los datos entre 0 y 1
scaler = MinMaxScaler(feature_range=(0,1))
data_esc = scaler.fit_transform(data)
```

```python
data_esc
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
iris_esc = scaler.fit_transform(X)
```

```python
print(X[:5])
print(iris_esc[:5])
```

```python
scaler2 = StandardScaler()
data_esc2 = scaler2.fit_transform(data)
data_esc2
```

```python
np.std(data_esc2)
```

```python
categorias = np.array([["rojo"], ["azul"], ["verde"], ["rojo"], ["azul"], ["verde"]])
```

```python
encoder = OneHotEncoder(sparse_output=False)
data_codificada = encoder.fit_transform(categorias)
data_codificada
```
