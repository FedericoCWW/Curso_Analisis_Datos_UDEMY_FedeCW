```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\Usuario\Documents\COSAS PYTHON DATA SCIENCE Y ML UDEMY)\datos_seguro.csv")
df
```

```python
plt.scatter(df.edad, df.compra);
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["edad"]], df["compra"], train_size=0.9)
```

```python
X_train
```

```python
X_test
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_train,y_train)
```

```python
modelo.score(X_train,y_train)
```

```python
datos_nuevos = pd.DataFrame({"edad": [25,35,55]})
```

```python
datos_nuevos
```

```python
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades
# probabilidades de que compre y de que no compre, lo vamos a separar
```

```python
probabilidades = probabilidades[:, 1]
```

```python
probabilidades
```

```python
plt.scatter(df.edad, df.compra)
plt.scatter(datos_nuevos["edad"], probabilidades);
```

```python
modelo.fit(X_train, y_train)
```

```python
score = modelo.score(X_test, y_test)
score
```
