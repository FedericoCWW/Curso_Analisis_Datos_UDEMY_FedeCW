```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(r"C:\Users\Usuario\Documents\COSAS PYTHON DATA SCIENCE Y ML UDEMY)\Ventas_dia11.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["Mes"] = df["Fecha"].dt.month
df["Dia"] = df["Fecha"].dt.day
df["FindeSemana"] = df["DíaDeLaSemana"].apply(lambda x: 1 if x >= 6 else 0)
```

```python
df
```

```python
# Escalar los datos
escala = MinMaxScaler(feature_range=(0,1))
```

```python
cols_para_escalar = df.drop(["Ventas", "Fecha"], axis=1).columns
```

```python
normado = escala.fit_transform(df[cols_para_escalar])
df_normado = pd.DataFrame(data=normado, columns=cols_para_escalar)
df_normado["Ventas"] = df["Ventas"]
df_normado["Fecha"] = df["Fecha"]
```

```python
df_normado
```

```python
X, y = df_normado.drop(["Ventas", "Fecha"], axis=1), df["Ventas"]
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df["Ventas"], kde=True, bins=30)
plt.title("Distribucion de ventas")
plt.xlabel("Ventas")
plt.ylabel("Cantidad");
```

```python
plt.figure(figsize=(12,5))
sns.barplot(df, x="DíaDeLaSemana", y="Ventas")
plt.title("Distribucion de ventas")
plt.xlabel("Ventas")
plt.ylabel("Cantidad")
plt.show();
```

```python
fig, axes = plt.subplots(3, 3, figsize=(30, 20), dpi=100)

fig.suptitle('Análisis Exploratorio de Ventas Profundo', fontsize=20, fontweight='bold')

# 2.1 Serie temporal de ventas
axes[0, 0].plot(df['Fecha'], df['Ventas'], linewidth=1)
axes[0, 0].set_title('Serie Temporal de Ventas (2022)')
axes[0, 0].set_xlabel('Fecha')
axes[0, 0].set_ylabel('Ventas')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2.2 Distribución de ventas
axes[0, 1].hist(df['Ventas'], bins=30, edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Distribución de Ventas')
axes[0, 1].set_xlabel('Ventas')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].axvline(df['Ventas'].mean(), color='red', linestyle='--', label=f'Media: {df["Ventas"].mean():.0f}')
axes[0, 1].legend()

# 2.3 Ventas por día de la semana
ventas_por_dia = df.groupby('DíaDeLaSemana')['Ventas'].mean().reset_index()
axes[0, 2].bar(ventas_por_dia['DíaDeLaSemana'], ventas_por_dia['Ventas'])
axes[0, 2].set_title('Ventas Promedio por Día de la Semana')
axes[0, 2].set_xlabel('Día de la Semana (1=Lunes, 7=Domingo)')
axes[0, 2].set_ylabel('Ventas Promedio')
axes[0, 2].set_xticks(range(1, 8))

# 2.4 Ventas: Festivos vs No Festivos
ventas_festivo = df.groupby('Festivo')['Ventas'].mean()
axes[1, 0].bar(['No Festivo', 'Festivo'], ventas_festivo.values, color=['blue', 'orange'])
axes[1, 0].set_title('Ventas Promedio: Festivos vs No Festivos')
axes[1, 0].set_ylabel('Ventas Promedio')
for i, v in enumerate(ventas_festivo.values):
    axes[1, 0].text(i, v, f'{v:.0f}', ha='center', va='bottom')

# 2.5 Ventas: Con Promoción vs Sin Promoción
ventas_promocion = df.groupby('Promociones')['Ventas'].mean()
axes[1, 1].bar(['Sin Promoción', 'Con Promoción'], ventas_promocion.values, color=['gray', 'green'])
axes[1, 1].set_title('Ventas Promedio: Con vs Sin Promoción')
axes[1, 1].set_ylabel('Ventas Promedio')
for i, v in enumerate(ventas_promocion.values):
    axes[1, 1].text(i, v, f'{v:.0f}', ha='center', va='bottom')

# 2.6 Ventas por mes
ventas_por_mes = df.groupby('Mes')['Ventas'].mean().reset_index()
axes[1, 2].bar(ventas_por_mes['Mes'], ventas_por_mes['Ventas'])
axes[1, 2].set_title('Ventas Promedio por Mes')
axes[1, 2].set_xlabel('Mes')
axes[1, 2].set_ylabel('Ventas Promedio')
axes[1, 2].set_xticks(range(1, 13))

# 2.7 Boxplot de ventas por día de la semana
df_boxplot = df.copy()
df_boxplot['DíaDeLaSemana'] = df_boxplot['DíaDeLaSemana'].astype(str)
sns.boxplot(x='DíaDeLaSemana', y='Ventas', data=df_boxplot, ax=axes[2, 0])
axes[2, 0].set_title('Distribución de Ventas por Día de la Semana')
axes[2, 0].set_xlabel('Día de la Semana')
axes[2, 0].set_ylabel('Ventas')

# 2.8 Correlación entre variables numéricas
numeric_cols = ['Ventas', 'DíaDeLaSemana', 'Promociones', 'Festivo', 'Mes', 'Dia']
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, ax=axes[2, 1])
axes[2, 1].set_title('Matriz de Correlación')

# 2.9 Ventas acumuladas por mes
ventas_mensuales = df.groupby('Mes')['Ventas'].sum().reset_index()
ventas_mensuales['Mes'] = ventas_mensuales['Mes'].astype(str)
axes[2, 2].bar(ventas_mensuales['Mes'], ventas_mensuales['Ventas'])
axes[2, 2].set_title('Ventas Totales por Mes')
axes[2, 2].set_xlabel('Mes-Año')
axes[2, 2].set_ylabel('Ventas Totales')
axes[2, 2].tick_params(axis='x', rotation=45)

plt.tight_layout()
```

```python
# Definiur el uso del modelo
X, y = df_normado.drop(["Ventas", "Fecha"], axis=1), df["Ventas"]

X = df[["DíaDeLaSemana", "Promociones", "Festivo", "Mes", "Dia", "FindeSemana"]]
y = df["Ventas"]

```

```python
#Uso el de regresion lineal para este caso y fue.
```

```python
#Definicion de la variable para el modelo
modelo = LinearRegression()
#Entrenaimeto del modelo con datos del dataset
modelo.fit(X_entrena, y_entrena)
```

```python
modelo.score(X_prueba, y_prueba)
```

```python
# Prediccion del modelo
prediccion = modelo.predict(X_prueba)
```

```python
plt.scatter(y_prueba, prediccion, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
```

```python
df_final =pd.DataFrame({"Real": y_prueba, "Predichas": prediccion})

plt.figure(figsize=(15,5))
plt.hist(df_final["Real"])
plt.hist(df_final["Predichas"], alpha=0.6, edgecolor='black', linewidth=1.5)

plt.show();
```
