```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
```

```python
df
```

```python
df.species.unique()
```

```python
X = df.drop("species", axis=1)
```

```python
X
```

```python
#labelecnoding es para traducir str a terreno de indice numericos.
le = LabelEncoder()
species = le.fit_transform(df["species"])
species
```

```python
y = species
y
```

```python
# ya tenemos x , y. Separamos registros en grupo de entranimiento y testeo
X_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
tree = DecisionTreeClassifier()
```

```python
tree
```

```python
tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=[20,20])
plot_tree(decision_tree=tree)
# representacion grafica de arbol
```

```python
# parte 2
```

```python
plt.figure(figsize=[20,20])
plot_tree(decision_tree=tree, filled=True, class_names=["Setosa", "Versicolor", "Virginica"], feature_names=df.columns.to_list())
```
