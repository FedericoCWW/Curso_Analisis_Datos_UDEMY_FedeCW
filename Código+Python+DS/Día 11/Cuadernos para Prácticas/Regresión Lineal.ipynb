```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

```python
graph = plt.scatter(df["Area"], df["Precio"]);
```

```python
X = df[["Area"]]
```

```python
X
```

```python
y = df["Precio"]
y
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
modelo.predict([[3300]])
```

```python
plt.plot(df["Area"], modelo.predict(X))
plt.scatter(df["Area"], df["Precio"], c="orange")
```
