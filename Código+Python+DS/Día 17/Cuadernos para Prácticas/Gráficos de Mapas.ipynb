```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(go.Scattergeo(lon=[-3.7, -78,2, -91.1],lat=[40.4, 19.4, -12.0], text=["Ciudad 1", "Ciudad 2", "Ciudad 3"], mode='markers')) 
fig.update_layout(title="Mapa de coordenadas", geo_scope='world')
```

```python
fig2 = go.Figure(go.Scattergeo(lon=[-3.7, -78,2, -21.1],lat=[40.4, 19.4, -12.0], text=["Ciudad 1", "Ciudad 2", "Ciudad 3"], mode='markers', marker=dict(size=[16,2,44], color=[10,20,30], colorscale="Turbo", showscale=True))) 
fig2.update_layout(title="Mapa de coordenadas", geo_scope='world')
```
