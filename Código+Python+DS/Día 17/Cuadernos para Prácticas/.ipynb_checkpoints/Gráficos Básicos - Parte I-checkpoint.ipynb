```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[10,15,13,17,14], mode='lines', name='Linea'))
```

```python
fig.update_layout(title="Grafio de Lineas", xaxis_title="X", yaxis_title="Y")
```

```python
fig2 = go.Figure()
```

```python
fig2.add_trace(go.Bar(x=[1,2,3,4,5], y=[5,6,7,8,9], marker=dict(color="yellowgreen", line=dict(color="green", width=1.5))))
```

```python
fig2.update_layout(title="Grafico de barras")
```

```python
fig3 = go.Figure()
```

```python
fig3.add_trace(go.Scatter(x=[1,2,3,4,5], y=[10,15,13,17,14], mode='lines', fill="tozeroy",name='area'))
```

```python
fig4 = go.Figure()
```

```python
fig4.add_trace(go.Pie(labels=['A', 'B', 'C', 'D'], values=[155,266,365,489]))
fig4.update_layout(title="Grafico de Torta")
```

```python
data = go.Histogram(x=[1,2,2,3,4,3])
```

```python
fig5 = go.Figure(data=data)
fig5.update_layout(title="Histograma")
```
