```python
import plotly.graph_objects as go
import numpy as np
```

```python
fig = go.Figure(data=[go.Scatter3d(x=[1,2,3,4], y=[5,6,7,8], z=[8,9,10,11], mode='markers')])
```

```python
fig
```

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))
```

```python
fig2 = go.Figure(data=[go.Surface(x=x, y=y, z=z)])
fig2
```
