```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

fig = make_subplots(rows=2, cols=2, subplot_titles=("Gráfico 1: Línea simple", 
                                                    "Gráfico 2: Línea con puntos", 
                                                    "Gráfico 3: Seno", 
                                                    "Gráfico 4: Coseno"))

x = list(range(10))

fig.add_trace(
    go.Scatter(x=x, y=[i**2 for i in x], mode='lines', name='y = x²'),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=x, y=[i**3 for i in x], mode='lines+markers', name='y = x³'),
    row=1, col=2
)

x_sin = np.linspace(0, 4*np.pi, 100)
fig.add_trace(
    go.Scatter(x=x_sin, y=np.sin(x_sin), mode='lines', name='sin(x)', line=dict(color='red')),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=x_sin, y=np.cos(x_sin), mode='lines', name='cos(x)', line=dict(color='green')),
    row=2, col=2
)

fig.update_layout(height=600, width=800, title_text="4 Gráficos Básicos", showlegend=False)
fig.update_xaxes(title_text="Eje X")
fig.update_yaxes(title_text="Eje Y")

fig.show()
```
