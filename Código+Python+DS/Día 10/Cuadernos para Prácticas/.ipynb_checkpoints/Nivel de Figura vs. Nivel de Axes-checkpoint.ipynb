```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos
```

```python
sns.displot(data=pinguinos, x="flipper_length_mm", hue="species", multiple="stack");
```

```python
sns.histplot(data=pinguinos, x="flipper_length_mm", hue="species", multiple="stack");
```

```python
sns.displot(data=pinguinos, x="flipper_length_mm", hue="species", col="species");
```

```python
graph = sns.relplot(data=pinguinos, x="flipper_length_mm", y="bill_length_mm", col="species")
graph.set_xlabels("Longitud del ala (mm)")
graph.set_ylabels("Longitud del pico (mm)");
```
