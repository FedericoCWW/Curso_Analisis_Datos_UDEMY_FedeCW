```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
```

```python
propinas
```

```python
sns.catplot(
    data=propinas,
    x="day",
    y="tip",
    kind="strip",
    hue="smoker"
);
```

```python
sns.catplot(
    data=propinas,
    x="day",
    y="tip",
    kind="strip",
    hue="smoker"
);
```
