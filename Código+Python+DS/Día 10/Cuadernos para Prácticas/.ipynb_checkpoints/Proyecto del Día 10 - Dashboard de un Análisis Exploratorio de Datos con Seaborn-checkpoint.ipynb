```python
import seaborn as sns
import pandas as pd
```

```python
# 1. importar dataset
diamantes = sns.load_dataset("diamonds")
diamantes
```

```python
#2. informacion basica del dataset on Pandas
diamantes.describe()
```

```python
diamantes.info()
```

```python
# Relaciones entre variables numéricas (por ejemplo, precio vs. peso en quilates)

sns.set(style="whitegrid")
g1 = sns.relplot(data=diamantes,
           x="carat",
           y="price",
           hue="cut",
           kind="scatter",
           height=6,
           aspect=1.5);
```

```python
#Distribuciones de variables numéricas (por ejemplo, precio, peso en quilates)
diamantes["peso"] = diamantes["carat"] / (0.2)

g2 = sns.displot(
    data=diamantes,
    x="peso",
    kind="hist",
    kde=True,
    height=4,
    aspect=2
)


g2.fig.suptitle("Distribucion del peso en quilates de los diamantes")
g2.set_axis_labels("Peso", "frequencia")
```
