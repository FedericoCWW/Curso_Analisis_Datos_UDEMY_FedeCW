## Consignas del Proyecto

5. Realizar un análisis simple de los datos para identificar tendencias o patrones que podrían ser útiles para decisiones de negocios o marketing.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
import uuid
import hashlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib
```

```python
ruta = r"H:\Cosas Fede\Proyectos_Desarrollo_2026\Curso_Analisis_Datos_UDEMY_FedeCW\clientes.csv"
df = pd.read_csv(ruta)
```

```python
df
```

```python
df['direccion'] = np.random.permutation(df['direccion'])
df
# le hice un shuffle a la direccion, lo mezcla pero no pierde la logica.
```

```python
df['edad'] = (df['edad'] // 10) * 10
```

```python
df
#converti datos a a que terminen en unidad '0'
```

```python
noise = np.random.normal(0, 100, size=df['salario'].shape)
df['salario'] *= noise
df['salario'] = np.abs(df['salario'] // 1000)
df

#reconverti a datos usando noise como aleatoredad y despues lo dividi por 1000 pasando por la funcion abs(valor absoluto de numpy)
```

```python
# Aplicar técnicas de pseudonimización al campo `nombre`.
lista_nombres = []
```

```python
lista_nombres.extend(map(lambda _: str(uuid.uuid4()), range(len(df))))
len(lista_nombres)
```

```python
lista_nombres
```

```python
df['Nombre_Pseudo'] = lista_nombres
df.drop('nombre', axis=1, inplace=True)
df
# Use tecnica de Hashing para el nombre con uuid4
```

```python
# 4. Implementar una función de balanceo de datos para asegurar la representación equitativa de diversas categorías en el análisis posterior.
def Datos_Balanceados_Cat(data):
    agrupado = data.groupby('categoria')
    data_balanceada = pd.DataFrame()
    for nombre, grupo in agrupado:
        grupo_balanceado = resample(grupo, 
                                   replace=True,
                                   n_samples=10,
                                   random_state=123)
        data_balanceada = pd.concat([data_balanceada, grupo_balanceado])
    return data_balanceada
        
```

```python
Datos_Balanceados_Cat(df)
```

```python
# 5. Realizar un análisis simple de los datos para identificar tendencias o patrones que podrían ser útiles para decisiones de negocios o marketing.
plt.figure(figsize=(12,8))
ax = sns.histplot(df['edad'], kde=True, color="orange", bins=30)
ax.set(title="Histograma por edades", xlabel='Edad', ylabel='Freq (Q)')
```

```python
ax = sns.scatterplot(x=df['edad'], y=df['salario'])
ax.set(title="Histograma por edades", xlabel='Edad', ylabel='Freq (Q)')
```

```python
plt.figure(figsize=(14, 8))

boxplot = sns.boxplot(
    y=df['edad'], 
    x=df['categoria'],
    palette='Set2',  # Colores más atractivos
    width=0.6,  # Ancho de las cajas
    linewidth=2  # Grosor de las líneas
)

plt.title('Distribución de Edad por Categoría', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Categoría', fontsize=12, fontweight='bold')
plt.ylabel('Edad', fontsize=12, fontweight='bold')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
